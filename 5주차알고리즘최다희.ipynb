```python
## 2156번문제

import sys
sys.setrecursionlimit(10**8) # 재귀 에러나면 재귀 깊이 늘려주기
n = int(input())
value = [0] 
dp = [0]*(n+1)

for _ in range(n):
    num = int(input())
    value.append(num)
    
def dpdp(cur): # 현재 위치에서 max return
    if cur < n+1 :
        dp[cur] = max(dp[cur-2]+value[cur], dp[cur-1], dp[cur-3]+value[cur-1]+value[cur])
        dpdp(cur+1)

dpdp(1)
print(max(dp))
```

```python
## 2293
from itertools import combinations_with_replacement

n, k = map(int,input().split())
arr = [] # 동전가치
pp = []

for _ in range(n):
    tmp = int(input())
    arr.append(tmp)

rmax = int(k/min(arr))+1
rmin = int(k/max(arr))

for i in range(rmin,rmax):
    pp += list(combinations_with_replacement(arr,i))
    
res = 0
for i in pp :
    if sum(i) == k :
        res +=1
print(res)
```

```python
import sys
sys.setrecursionlimit(10**8) # 재귀 에러나면 재귀 깊이 늘려주기
#input = sys.stdin.readline
n, k = map(int,input().split())
arr = [] # 동전가치

for _ in range(n):
    tmp = int(input())
    arr.append(tmp)
    
pp = [0]*(k+1)

def cnt(pp,cur,index,k):
    for i in arr :
        if cur <= i :
            if index+i <= k :
                pp[index+i] += 1
                cnt(pp,i,index+i,k)
            
cnt(pp,min(arr),0,k)
print(pp[k])
```

```python
n, k = map(int, input().split())
arr = []
for _ in range(n):
    arr.append(int(input()))
dp = [0 for _ in range(k+1)]

dp[0] = 1
for c in arr:
    for i in range(k+1):
        if i - c >= 0:
            dp[i] += dp[i-c]

print(dp[k])
```
