### Additive Feature Attribution Methods
-----------------
- 단순한 모델에 대한 가장 좋은 설명은 모델 자체이지만 앙상블 혹은 딥러닝의 경우 원본 모델을 가장 좋은 설명으로 사용할 수 없음
- 원래 모델의 해석 가능한 근사치로 정의하는 더 간단한 설명 모델을 사용해야 함

**2.1 LIME**

입력 데이터의 변형을 가했을 때 모델 예측에 어떠한 변화가 일어나는지 테스트

관심있는 관측치(인스턴스)에 대한 샘플링된 관측치의 근접성(proximity)에 의해 가중치가 부여된 해석 가능한 모델을 학습

**2.2 DeepLIFT**

reference를 설정하여 비교를 통해 변수중요도를 산출하고, Positive / Negative contribution에 대해 개별적인 방법을 적용

**2.3 Layer-Wise Relevance Propagation**

딥러닝 모델의 결과를 역추적해서 입력 이미지에 히트맵을 출력

**2.4 Classic Shapley Value Estimation**

Shapley regerssion value ; 다중 공선성이 있는 선형 모형에 대한 특성 중요성, 다중공선성은 선형 모델에서 독립성 가정을 위배하는 성질이지만, Shapley regression value는 다중공선성이 있는 선형 모델에서도 사용할 수 있는 값. Shapley regression value는 각 변수들이 학습에 포함되었을 때, 얼마나 모델의 성능에 영향을 미치는지에 따라 중요도(importance value)를 부여함

(중요도 계산 방식)

###  SHAP (SHapley Additive exPlanation) Values
-----------------

- SHAP values를 특징 중요도의 통일된 척도로 제안함

Linear SHAP

선형 모델의 경우 입력 형상 독립성(식 11)을 가정하면 SHAP 값은 모델의 가중치 계수에서 직접 근사할 수 있습니다.

Max SHA

Shapley 값의 순열 공식을 사용하여 각 입력이 다른 모든 입력에 대해 최대값을 증가시킬 확률을 계산할 수 있습니다. 입력 값의 정렬 순서에 따라 이 작업을 수행하면 O(M2^M) 시간이 아닌 O(M^2) 시간 내에 M 입력이 있는 최대 함수의 Shapley 값을 계산할 수 있습니다. 

Deep SHAP (DeepLIFT + Shapley values)
